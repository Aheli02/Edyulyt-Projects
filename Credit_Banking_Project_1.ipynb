```python
import numpy as np
import pandas as pd
```

```python
#Loading the Dataset
Customer_acqusition = pd.read_csv('/content/Customer_Acqusition - Customer Acqusition.csv')
Repayment = pd.read_csv('/content/Repayment - Repayment.csv')
Spend = pd.read_csv('/content/Spend - Spend.csv')
```

**Sanity Checks - Data Cleaning**

```python
#Replacing ages less than 18 with NaN
Customer_acqusition['Age'] = Customer_acqusition['Age'].mask(Customer_acqusition['Age'] < 18)
```

```python
Customer_acqusition
```

```python
Customer_acqusition.to_excel("Customer_acqusition.xlsx")
```

```python
#Renaming 'Amount' columns for 'Repayment' and 'Spend' dataframe
Repayment.rename(columns={'Amount': 'Amount_Repayment','Costomer': 'Customer' }, inplace=True)
Spend.rename(columns={'Amount': 'Amount_Spend', 'Costomer': 'Customer'}, inplace=True)
```

```python
Repayment
```

```python
Repayment.to_excel("Repayment.xlsx")
```

```python
Spend
```

```python
Spend.to_excel("Spend.xlsx")
```

```python
#Giving credio of 2% of the surplus amount to customers whose repayment
#amount is more than their spend amount, in the next month billing.
Merged = pd.merge(Repayment, Spend, on=['Customer', 'Month'])
Merged['Surplus'] = Merged['Amount_Repayment'] - Merged['Amount_Spend']
```

```python
Merged
```

```python
Next_Month_Billing = Merged[Merged['Surplus'] > 0]
Next_Month_Billing['Credit'] = 0.02 * Next_Month_Billing['Surplus']
Next_Month_Billing['Next Month Billing'] = Next_Month_Billing['Amount_Spend'] - Next_Month_Billing['Credit']
```

```python
Next_Month_Billing
```

```python
Next_Month_Billing.to_excel("Next_Month_Billing.xlsx")
```

**Tasks:**

```python
#Calculating monthly spend of each customer
Monthly_Spend = Spend.groupby(['Customer', 'Month'])['Amount_Spend'].sum()
print(Monthly_Spend)
```

```python
Monthly_Spend.to_excel("Task_1.xlsx")
```

```python
#Calculating the monthly repayment of each customer
Monthly_Repayment = Repayment.groupby(['Customer', 'Month'])['Amount_Repayment'].sum()
print(Monthly_Repayment)
```

```python
Monthly_Repayment.to_excel("Task_2.xlsx")
```

```python
#Calculating 10 highest paying customers
Sorted = Repayment.sort_values(by='Amount_Repayment', ascending=False)
Highest_Paying_Customers = Sorted.head(10)

print("The top 10 highest paying customers:-")
print(Highest_Paying_Customers)
```

```python
Highest_Paying_Customers.to_excel("Task_3.xlsx")
```

```python
#Determining segment of customers spending more money
Merged_2 = pd.merge(Customer_acqusition, Spend, on='Customer')

Segment_Spending = Merged_2.groupby('Segment')['Amount_Spend'].sum()
Sorted_Segments = Segment_Spending.sort_values(ascending=False)
print("Decreasing order of segments according to spend :")
print(Sorted_Segments)
```

```python
Sorted_Segments.to_excel("Task_4.xlsx")
```

```python
#Determining which age group is spending more money
Age_Group_Spending = Merged_2.groupby('Age')['Amount_Spend'].sum()
Sorted_Age_Group = Age_Group_Spending.sort_values(ascending=False)
print("Decreasing order of age groups according to spend :")
print(Sorted_Age_Group)
```

```python
Sorted_Age_Group.to_excel("Task_5.xlsx")
```

```python
#Determining the most profitable segment
Merged_3 = pd.merge(Customer_acqusition, Repayment, on='Customer')
Segment_Repaying = Merged_3.groupby('Segment')['Amount_Repayment'].sum()
Segment_Profit = Segment_Spending - Segment_Repaying

Most_Profitable_Segment = Segment_Profit.idxmax()
print("The Most profitable segment:", Most_Profitable_Segment)
```

```python
#Determining in which category the customers are spending more money
Category_Spending = Spend.groupby('Type')['Amount_Spend'].sum()
Highest_Spending_Category = Category_Spending.idxmax()
print("Category with the highest spending:", Highest_Spending_Category)
```
